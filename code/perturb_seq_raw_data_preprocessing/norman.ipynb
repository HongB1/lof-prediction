```python
import pandas as pd

def load_info(dir_path, geo_id, filtered, load_matrix=False):

    barcode_path = f'{dir_path}/{geo_id}_{filtered}_barcodes.tsv'
    cell_identites = f'{dir_path}/{geo_id}_{filtered}_cell_identities.csv'
    genes = f'{dir_path}/{geo_id}_{filtered}_genes.tsv'
    matrix = f'{dir_path}/{geo_id}_{filtered}_matrix.mtx'

    barcode = pd.read_csv(barcode_path, sep='\t', header=None, names=['cell_barcode'])
    cell_identities = pd.read_csv(cell_identites, sep=',', header=None)
    genes = pd.read_csv(genes, sep='\t', header=None)
    if load_matrix:
        matrix = scanpy.read_mtx(matrix)
    else:
        matrix = None
    return barcode, cell_identities, genes, matrix
```
