```python
exp = anndata.read_mtx(f"/media/hdd/hb/lof/perturb_seq/geo_raw_data/replogle_gse146194/{gsm_id}.matrix.mtx")
exp
```

```python
barcode  = pd.read_csv(f"/media/hdd/hb/lof/perturb_seq/geo_raw_data/replogle_gse146194/{gsm_id}.barcodes.tsv", sep='\t', header=None, names=['cell_barcode'])
barcode.index = barcode['cell_barcode']
del barcode['cell_barcode']

cell_info = pd.read_csv(f"/media/hdd/hb/lof/perturb_seq/geo_raw_data/replogle_gse146194/{gsm_id}.raw_cell_identities.csv",)
# cell_info.drop(columns=['guide_identity', 'A', 'B', 'protospacer_A', 'protospacer_B'], inplace=True)
cell_info.index = cell_info['cell_barcode']
del cell_info['cell_barcode']

gene_info = pd.read_csv(f"/media/hdd/hb/lof/perturb_seq/geo_raw_data/replogle_gse146194/{gsm_id}.features.tsv", header=None, sep='\t',  names=['gene_id', 'gene_name', 'type' ])
gene_info.index = gene_info['gene_id']
del gene_info['gene_id']
```

```python
# good_coverage == True만 뽑기
cell_info_f = cell_info[cell_info['good_coverage']!=False]
cell_info_f.good_coverage.unique()
```

```python
# condition_list = []
# cell_info_f_condition = [x if x != 'Non-Targeting' else 'control' for x in cell_info_f.target ]
# cell_info_f.insert(0, 'condition', cell_info_f_condition)
# cell_info_f
cell_info_f['condition'] = [x for x in cell_info_f['target_A'] + '/' + cell_info_f['target_B'].values]
cell_info_f['condition'] = cell_info_f['condition'].str.replace('Non-Targeting/Non-Targeting', 'control')
cell_info_f['condition'] = cell_info_f['condition'].str.replace('Non-Targeting/', '')
cell_info_f['condition'] = cell_info_f['condition'].str.replace('/Non-Targeting', '')
cell_info_f
```

```python
exp = exp.to_df().T
```

```python
exp.obs = barcode
exp.var = gene_info
```

```python
exp.index = barcode.index
exp.columns = gene_info.index
```

```python
cell_info_f
```

```python
exp = exp[exp.index.isin(cell_info_f.index)]
cell_info_f = cell_info_f[cell_info_f.index.isin(exp.index)]
```

```python
exp.obs = cell_info_f
exp.var = gene_info
```

```python
exp.sort_index(inplace=True)
cell_info_f.sort_index(inplace=True)
```

```python
adata = anndata.AnnData(X=exp, obs=cell_info_f, var=gene_info)
```
